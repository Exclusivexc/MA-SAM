## Import augmnetation function and load image

```python
import pickle
import numpy as np
from augmentation import *

image = pickle.load(open('/media/ubuntu/maxiaochuan/MA-SAM/data/Promise12/2D_slices/Case00_slice03.pkl', 'rb'))
image = image.reshape(image.shape[0], image.shape[1])
def scale_image(image):
    return ((image - image.min()) / (image.max() - image.min()) * 255).astype(np.uint8)
```

## apply augmentation to the image

```python
print(image.shape, image.min(), image.max())
flipped_image, axis = random_flip(image)
rotated_image, angle = random_rotate(image)
input = (image - image.min()) / (image.max() - image.min())
gammaed_image, gamma = gamma_correction(input)
posterized_image, bits = posterization(input)
contrast_adjustment_image, factor = contrast_adjustment(input)
brightness_adjustment_image, factor = brightness_modification(input)
sharped_image, factor = sharpness_enhancement(input)
```
